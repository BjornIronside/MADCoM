```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dfs = []
for filename in os.listdir():
    if filename[-4:] != '.csv':
        continue
    
    file_df = pd.read_csv(filename, index_col=False)
    spl = filename[:-4].split('_')
    file_df['init_type'] = spl[1]
    file_df['instance'] = '_'.join(spl[2:])
    dfs.append(file_df)
df = pd.concat(dfs)
df.reset_index(inplace=True, drop=True)
df.head()
```

```python
instance_data = {
    'Beijing-6': 2561113,
    'Beijing-8': 3136727,
    'Hefei-10': 1748829,
    'K5_g-6': 3109473,
    'N5_g-5': 3325667,
    'O1_g-4': 2597731,
    'O1_p-2': 2146477,
    'S2_g-6': 1439140
}
df['gap'] = 0
df['total_time'] = df['ls_time'] + df['hd_time']
for i in df.index.values:
    df.loc[i, 'gap'] = df.loc[i, 'cost'] / instance_data[df.loc[i, 'instance']] - 1
df['gap'] = df['gap'] * 100
df.head()
```

```python
groups = df.groupby('init_type')
old = groups.get_group('old')
hd = groups.get_group('hd')
ten = groups.get_group('ten')
unlimited = groups.get_group('unlimited')
init_df = pd.concat([old, ten, hd, unlimited])
init_df.tail()
```

```python
#sns.violinplot(data=df, x='instance', y='gap', hue='init_type', ax=ax)
g = sns.catplot(data=init_df, x='init_type', y='gap', col='instance', kind='box', 
            col_wrap=4, sharey=False, whis=20)
g.set_axis_labels(y_var="Gap [%]", x_var="Initialization Method", fontsize=14)
g.set_yticklabels(fontsize=12)
g.set_xticklabels(['Classic','Max. 10',r'Max. $\sqrt{n}$', 'Unlimited'], fontsize=12)
g.set_titles("Instance {col_name}")
```

```python

#sns.boxplot(data=df, x='instance', y='diversity', hue='init_type', ax=ax)
g = sns.catplot(data=init_df, x='init_type', y='diversity', col='instance', kind='box', 
            col_wrap=4, sharey=False, whis=200)
g.set_axis_labels(y_var="Percentage of maximum diversity [%]", x_var="Initialization Method", fontsize=14)
g.set_yticklabels(fontsize=12)
g.set_xticklabels(['Classic','Max. 10',r'Max. $\sqrt{n}$', 'Unlimited'], fontsize=12)
g.set_titles("Instance {col_name}", fontsize=12)
```

```python
#fig, ax = plt.subplots(figsize=(16,10))
#sns.boxplot(data=df, x='instance', y='total_time', hue='init_type', ax=ax)
g = sns.catplot(data=init_df, x='init_type', y='total_time', col='instance', kind='box', 
            col_wrap=4, sharey=False, whis=20)
g.set_axis_labels(y_var="Solution Generation Time [s]", x_var="Initialization Method", fontsize=14)
g.set_yticklabels(fontsize=12)
g.set_xticklabels(['Classic','Max. 10',r'Max. $\sqrt{n}$', 'Unlimited'], fontsize=12)
g.set_titles("Instance {col_name}", fontsize=12)
```
